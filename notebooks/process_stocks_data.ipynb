```python
import pandas as pd
```

```python
stock_list_file = 'US_Stocks_list.xlsx'
data_file = 'US_STOCKS_MONTHLY.xlsx'
```

```python
tickers = list(pd.read_excel(stock_list_file, index_col=0).index.values)
```

```python
tickers
```

```python
len(tickers)
```

```python
def chunks(lst, n):
    """Yield successive n-sized chunks from lst."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
```

```python
pairs = []
for p in chunks(tickers, 2):
    pairs.append(p)
```

```python
pairs
```

```python
data = pd.read_excel(data_file, index_col=0)
```

```python
data.dropna()
```

```python
clean_df = data.dropna()
```

```python
for p in pairs:
    asset1 = p[0]
    asset2 = p[1]
    clean_df['ratio_{}_{}'.format(asset1, asset2)] = clean_df['{}_Close'.format(asset1)] / clean_df['{}_Close'.format(asset2)]
```

```python
ratio_cols = ['ratio_{}_{}'.format(p[0], p[1]) for p in pairs]
```

```python
clean_df[ratio_cols].diff()/clean_df[ratio_cols].shift()
```

```python
asset_returns = clean_df[ratio_cols].diff()/clean_df[ratio_cols].shift()
```

```python
clean_returns = asset_returns.dropna()
```
